함수만들기

```python
def oracle(sql):
    dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
    conn = cx_Oracle.connect('busan','dbdb', dsn)
    cursor = conn.cursor()
    
    cursor.execute(sql)
    
    global rows, cols # 전역변수
    rows = cursor.fetchall() # 관측값
    colnames = cursor.description # 변수정보
    cols = [colnames[i][0] for i in range(len(colnames))] # 변수명

    cursor.close()
    conn.close()
    
    return pd.DataFrame(rows, columns=cols)
```

```python
def Oracle_part(temp_select, temp_from, temp, temp_group=None, temp_order=None):
    dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
    conn = cx_Oracle.connect('busan','dbdb', dsn)
    cursor = conn.cursor()
    
    # Select
    sql1 = 'Select ' + temp_select
    
    # From
    sql1 += '\n From ' + temp_from
    
    # 조건문 처리
    for val_k, val_v in zip(list(temp.keys()), list(temp.values())):
        if val_v.split(' ')[0] not in ['>', '<', '>=', '<=', '=', 'like', 'Like']:
            if val_k == list(temp.keys())[0]:
                sql2 = '\n Where ' + val_k + " = '" + val_v + "'"
                sql1 += sql2
            else:
                sql2 = '\n and ' + val_k + " = '" + val_v + "'"
                sql1 += sql2
        else:
            if val_k == list(temp.keys())[0]:
                sql2 = '\n Where ' + val_k + ' ' + val_v
                sql1 += sql2
            else:
                sql2 = '\n and ' + val_k + ' ' + val_v
                sql1 += sql2
                
    # Group BY
    if temp_group != None:
        sql1 += '\n Group BY ' + temp_group
    
    # Order By
    if temp_order != None:
        sql1 += '\n Order By ' + temp_order
    
    cursor.execute(sql1)

    global rows, cols # 전역변수
    rows = cursor.fetchall() # 관측값
    colnames = cursor.description # 변수정보
    cols = [colnames[i][0] for i in range(len(colnames))] # 변수명

    cursor.close()
    conn.close()

    return pd.DataFrame(rows, columns=cols)
```

회원 거주지역별 주문수량의 합을 조회하기
- 정렬 : 거주지역 오름차순
- 상품분류명에 '전자제품'이 포함
- 거래처명에 '대우전자'가 포함된 경우
- 함수는 편한 것 사용
- df8로 저장

```python
sql = '''Select Distinct substr(mem_add1,1,2) as mem_add, sum(cart_qty) as qty_sum
        From member, cart, prod, lprod, buyer
        Where mem_id = cart_member
            and cart_prod = prod_id
            and prod_buyer = buyer_id
            and prod_lgu = lprod_gu
            and lprod_nm like '%{}%'
            and buyer_name like '%{}%'
        Group By substr(mem_add1,1,2)
        Order By mem_add Asc
'''.format('전자제품', '대우전자')
df8 = oracle(sql)
df8
```

```python
Oracle_part('Distinct substr(mem_add1,1,2) as mem_add, sum(cart_qty) as qty_sum',
           'member, cart, prod, lprod, buyer',
           {'mem_id' : '= cart_member',
            'cart_prod' : '= prod_id',
            'prod_buyer' : '= buyer_id',
            'prod_lgu' : '= lprod_gu',
           'lprod_nm' : "like '%전자제품%'",
           'buyer_name' : "like '%대우전자%'"},
           'substr(mem_add1,1,2)',
           'mem_add Asc')
```
