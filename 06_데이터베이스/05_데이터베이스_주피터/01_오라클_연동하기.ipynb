```python
import cx_Oracle
```

* 데이터베이스 연동순서
    - 모든 데이터베이스 시스템 공통
    1. DB 드라이버 연결
    2. DB접속
    3. 커서 생성
    4. SQL요청 및 응답결과 받아오기
    5. 결과 출력
    6. DB접속끊기(커서 반환, 접속종료)

```python
# 1. 드라이버 연결
dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
dsn
```

```python
# 2. DB접속
conn = cx_Oracle.connect('busan','dbdb', dsn)
conn
```

```python
# 3. 커서 생성 : 응답과 요청을 수행
# SQL 구문을 실어 오라클에게 전달하고 응답결과를 받아오는 역할수행
cursor = conn.cursor()
cursor
```

```python
# 4. SQL요청 및 응답결과 받아오기
## sql구문 작성
sql = 'select * From member'

## SQL구문을 커서에 실어 보내기(요청)
# - execute() : 요청함수
# 결과는 cursor가 받아서 가지고 있음
cursor.execute(sql)
```

```python
# 5. 결과 출력
cursor.fetchall()
```

```python
# 6. DB 접속끊기(커서반환, 접속종료)
# 받아올 때 : DB -> cursor
# 끊을 때 : cursor -> DB
cursor.close()
conn.close()
```
