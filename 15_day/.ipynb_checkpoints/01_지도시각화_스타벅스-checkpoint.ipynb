# 지도시각화

```python
# 지도 밑바탕
starbucks_map = folium.Map(
    # 최초에 보여줄 지도 중심위치 지정
    # 서울 중심점
    location=[37.573050, 126.979189], #35.171564, 129.131751
    
    # 지도 스타일 지정하기
    # 도시형 건물 스타일 지정(가장 일반적)
    titles = 'openstreetmap',
    
    # 최초에 보여줄 zoom사이즈 지정하기
    zoom_start = 50    
)
starbucks_map
```

```python
#마커 표시
for i in range(0, len(seoul_starbucks)):
    lat = seoul_starbucks.iloc[i, 1]  #위도
    lng = seoul_starbucks.iloc[i, 2]  #경도

    folium.CircleMarker(
        location=[lat, lng],  # 마커 위치
        # 마커 속성
        fill_color='black',  #색상
        color='yellow',  #테두리
        weight=1,  #테두리 두께
        radius=5  #크기
    ).add_to(starbucks_map)  #마커 표시
starbucks_map
```

# 매장 타입별 마커 색상과 사이즈 달리해 지도 표시
- 레이어형태로 만들어지기 때문에 초기화 후 리스타트

```python
seoul_starbucks['매장타입'].value_counts()
```

```python
# 지도 밑바탕
starbucks_map2 = folium.Map(
    location=[37.573050, 126.979189], #35.171564, 129.131751
    titles = 'openstreetmap',
    zoom_start = 50    
)
```

```python
#마커 표시
for i in range(0, len(seoul_starbucks)):
    lat = seoul_starbucks.iloc[i, 1]  #위도
    lng = seoul_starbucks.iloc[i, 2]  #경도
    store = seoul_starbucks.iloc[i, 3]
    
    if store == 'general':
        folium.CircleMarker(
            location=[lat, lng],  # 마커 위치
            fill_color='green',  #색상
            color='green',  #테두리
            weight=1,  #테두리 두께
            radius=3  #크기
        ).add_to(starbucks_map2)  #마커 표시
        
    elif store == 'reserve':
        folium.CircleMarker(
            location=[lat, lng],  # 마커 위치
            fill_color='blue',  #색상
            color='blue',  #테두리
            weight=1,  #테두리 두께
            radius=5  #크기
        ).add_to(starbucks_map2)  #마커 표시
        
    else:
        folium.CircleMarker(
            location=[lat, lng],  # 마커 위치
            fill_color='red',  #색상
            color='red',  #테두리
            weight=1,  #테두리 두께
            radius=5  #크기
        ).add_to(starbucks_map2)  #마커 표시
starbucks_map2
```

```python
#강사님.ver
for i in range(0, len(seoul_starbucks)):
    lat = seoul_starbucks.iloc[i, 1]  #위도
    lng = seoul_starbucks.iloc[i, 2]  #경도
    store = seoul_starbucks.iloc[i, 3]
    
    if store == 'general':
        f_color = 'green'
        f_size = 3
        
    elif store =='reserve':
        f_color = 'blue'
        f_size = 5
        
    else:
        f_color = 'red'
        f_size = 5
        
    folium.CircleMarker(
        location=[lat, lng],  # 마커 위치
        fill_color=f_color,  #색상
        color=f_color,  #테두리
        weight=1,  #테두리 두께
        radius=f_size  #크기
    ).add_to(starbucks_map2)  #마커 표시
starbucks_map2
```
